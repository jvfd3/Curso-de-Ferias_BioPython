```python
from Bio.Seq import Seq
my_seq = Seq("GATCG")
for index, letter in enumerate(my_seq):
    print("%i %s" % (index, letter))
```

```python
# Cálculo manual de GC% (lembrar para depois)
my_seq2 = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
print("\nTamanho da sequência longa:", len(my_seq2))
print("Número de G:", my_seq2.count("G"))
gc_manual = 100 * (my_seq2.count("G") + my_seq2.count("C")) / len(my_seq2)
print("GC% manual:", gc_manual)
```

```python
#Fatiar sequência
my_seq3 = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
print("\nFatiamento simples (posição 4 a 11):", my_seq3[4:12])
print("Segunda base de cada códon:", my_seq3[0::3])
print("Terceira base de cada códon:", my_seq3[1::3])
print("Sequência invertida:", my_seq3[::-1])
```

Uso de Seq com funções biopython

NA ➜ RNA ➜ Proteína



```python
#Cálculo de GC%

from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
my_seq2 = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
gc_fraction(my_seq2)
```

```python
from Bio.Seq import Seq

dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")

print(dna)                     # sequência original
print(dna.complement())        # fita complementar
print(dna.reverse_complement())# reverso complementar
print(dna.transcribe())        # DNA -> RNA
print(dna.translate())         # DNA -> proteína
```

```python
#Fita codificadora e Fita Molde

from Bio.Data import CodonTable

coding_dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
print("Fita codificadora (DNA):", coding_dna)

# A fita molde é o complemento reverso
template_dna = coding_dna.reverse_complement()
print("Fita molde (DNA):       ", template_dna)
```

```python
#Transcrição DNA -> RNA

# Transcrição "bioinformática" (da fita codificadora)
# Apenas substitui T por U
messenger_rna = coding_dna.transcribe()
print("\nmRNA transcrito da fita codificadora:", messenger_rna)

# Transcrição biológica real
real_mrna = messenger_rna.back_transcribe()
print("mRNA a partir da fita molde:         ", real_mrna)
```

```python
#Tradução
print("\nTradução do mRNA:")
protein_from_rna = messenger_rna.translate()
print(protein_from_rna)

print("\nTradução direto do DNA codificador:")
protein_from_dna = coding_dna.translate()
print(protein_from_dna)
```

```python
#Tabela de tradução

standard_table = CodonTable.unambiguous_dna_by_name["Standard"]
print(standard_table)
```

```python
#Paradas biológicas

print("\nTradução padrão:")
print(coding_dna.translate())

print("Tradução até o primeiro códon STOP:")
print(coding_dna.translate(to_stop=True))

# Alterando símbolo de parada
print("Símbolo de parada personalizado:")
print(coding_dna.translate(table=2, stop_symbol="@"))
```

```python
#Tradução de CDS

# CDS = começa com códon start válido, termina com stop,
# não tem stops internos e comprimento múltiplo de 3

gene = Seq(
    "GTGAAAAAGATGCAATCTATCGTACTCGCACTTTCCCTGGTTCTGGTCGCTCCCATGGCA"
    "GCACAGGCTGCGGAAATTACGTTAGTCCCGTCAGTAAAATTACAGATAGGCGATCGTGAT"
    "AATCGTGGCTATTACTGGGATGGAGGTCACTGGCGCGACCACGGCTGGTGGAAACAACAT"
    "TATGAATGGCGAGGCAATCGCTGGCACCTACACGGACCGCCGCCACCGCCGCGCCACCAT"
    "AAGAAAGCTCCTCATGATCATCACGGCGGTCATGGTCCAGGCAAACATCACCGCTAA"
)

print("\nTradução bacteriana normal:")
print(gene.translate(table="Bacterial")[:60], "...")

print("Tradução bacteriana até STOP:")
print(gene.translate(table="Bacterial", to_stop=True)[:60], "...")

# Aqui o GTG é códon start alternativo → vira Metionina (M)
print("Tradução como CDS completa (start alternativo tratado):")
print(gene.translate(table="Bacterial", cds=True)[:60], "...")
```

```python
from Bio.Seq import Seq, MutableSeq
from Bio.Seq import UndefinedSequenceError

#Sequência com conteúdo desconhecido
# Sabemos o comprimento, mas NÃO sabemos as letras

unknown_seq = Seq(None, 10)

print("Sequência desconhecida:", unknown_seq)
print("Comprimento conhecido:", len(unknown_seq))

# Tentar acessar o conteúdo gera erro
try:
    print(unknown_seq[0])
except UndefinedSequenceError as e:
    print("Erro ao acessar conteúdo:", e)
```
